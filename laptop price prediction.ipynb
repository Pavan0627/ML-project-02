```python
# Function to extract memory type and capacity
def extract_memory_type(memory_str):
    if 'SSD' in memory_str:
        return 'SSD'
    elif 'HDD' in memory_str:
        return 'HDD'
    elif 'Hybrid' in memory_str:
        return 'Hybrid'
    elif 'Flash Storage' in memory_str:
        return 'Flash Storage'
    else:
        return None

# Apply the function to the 'memory' column to create a new column 'Memory Type'
lap['Memory Type'] = lap['Memory'].apply(extract_memory_type)

# Function to extract the capacity value
def extract_capacity(memory_str):
    memory_str = memory_str.replace('GB', '').replace('TB', '').strip()
    if 'SSD' in memory_str or 'HDD' in memory_str or 'Hybrid' in memory_str or 'Flash Storage' in memory_str:
        if 'TB' in memory_str:
            capacity_in_bytes = float(memory_str.split()[0]) * 1024 * 1024  # Convert TB to GB (1TB = 1024GB)
        else:
            capacity_in_bytes = float(memory_str.split()[0]) * 1024  # Convert GB to MB (1GB = 1024MB)
        return capacity_in_bytes
    else:
        return None

# Apply the function to the 'memory' column to create a new column 'Capacity'
lap['Capacity'] = lap['Memory'].apply(extract_capacity)

# Pivot the DataFrame to create separate columns for each memory type
memory_types = lap.pivot(columns='Memory Type', values='Capacity')

# Rename columns to match the desired names
memory_types.rename(columns={'SSD': 'SSD', 'HDD': 'HDD', 'Hybrid': 'Hybrid', 'Flash Storage': 'Flash Storage'}, inplace=True)

# Fill missing values with 0 for better representation
memory_types.fillna(0, inplace=True)

# Combine the original DataFrame with the new memory type columns
lap = pd.concat([lap, memory_types], axis=1)

# Drop the original 'memory' column, as we have extracted its information into new columns
lap.drop('Memory', axis=1, inplace=True)

# Convert the new columns to float data type
lap['SSD'] = lap['SSD'].astype('float')
lap['HDD'] = lap['HDD'].astype('float')
lap['Hybrid'] = lap['Hybrid'].astype('float')
lap['Flash Storage'] = lap['Flash Storage'].astype('float')
```

It seems like you're importing various modules from the scikit-learn library and the XGBoost library. Here's a breakdown of what each of these modules is used for:

>sklearn.compose.ColumnTransformer: This module is used for combining multiple feature extraction or transformation techniques into a single transformer. It allows you to specify different transformations for different columns in your dataset.

>sklearn.pipeline.Pipeline: This module is used for creating a sequence of data processing steps, where each step is a transformer or an estimator. It simplifies the workflow by encapsulating multiple preprocessing and modeling steps into a single object.

>sklearn.preprocessing.OneHotEncoder: This module is used for converting categorical features into a numerical representation using one-hot encoding. It creates binary columns for each unique category in the categorical variable.

>sklearn.metrics.r2_score: This module provides the R-squared score, which is a statistical measure indicating the proportion of the variance in the dependent variable that can be explained by the independent variables. It is commonly used to evaluate regression models.

>sklearn.metrics.mean_absolute_error: This module provides the mean absolute error, which measures the average absolute difference between the predicted and actual values in a regression task. It is another evaluation metric for regression models.

>sklearn.linear_model.LinearRegression: This module implements ordinary least squares linear regression, which fits a linear model to minimize the residual sum of squares between the observed and predicted values.

>sklearn.linear_model.Ridge: This module implements ridge regression, which adds a penalty term to the ordinary least squares objective function to control the complexity of the model.

>sklearn.linear_model.Lasso: This module implements Lasso regression, which also adds a penalty term to the objective function but uses the L1 norm penalty, allowing for feature selection by shrinking some coefficients to zero.

>sklearn.neighbors.KNeighborsRegressor: This module implements k-nearest neighbors regression, which predicts the target value for a sample by averaging the target values of its k nearest neighbors.

>sklearn.tree.DecisionTreeRegressor: This module implements a decision tree-based regression algorithm. It partitions the feature space into regions and predicts the target value based on the average value of the samples in each region.

>sklearn.ensemble.RandomForestRegressor: This module implements the random forest regression algorithm, which builds an ensemble of decision trees and averages their predictions to make the final prediction.

>sklearn.ensemble.GradientBoostingRegressor: This module implements gradient boosting regression, which builds an ensemble of weak regression models and sequentially improves their predictions by minimizing a loss function.

>sklearn.ensemble.AdaBoostRegressor: This module implements AdaBoost regression, which combines multiple weak regression models to create a strong model by giving more weight to samples with larger errors.

>sklearn.ensemble.ExtraTreesRegressor: This module implements the extremely randomized trees regression algorithm, which is similar to random forests but uses random thresholds for each feature to split nodes, resulting in higher randomness.

>sklearn.svm.SVR: This module implements support vector regression, which uses support vector machines to perform regression tasks. It finds the hyperplane that maximally separates the target values while keeping a tolerance zone.

>xgboost.XGBRegressor: This module provides an implementation of the XGBoost regression algorithm, which is an optimized gradient boosting algorithm known for its performance and flexibility.

```python
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score,mean_squared_error, r2_score,mean_absolute_error
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor, ExtraTreesRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
from sklearn.base import TransformerMixin, BaseEstimator
```

```python
# Custom Transformer class for regression models
class RegressorTransformer(TransformerMixin, BaseEstimator):
    def __init__(self, model):
        self.model = model

    def fit(self, X, y):
        self.model.fit(X, y)
        return self

    def transform(self, X):
        return self.model.predict(X)

# Create instances of each algorithm
linear_reg = LinearRegression()
ridge = Ridge()
lasso = Lasso()
knn = KNeighborsRegressor()
decision_tree = DecisionTreeRegressor()
random_forest = RandomForestRegressor()
gradient_boosting = GradientBoostingRegressor()
adaboost = AdaBoostRegressor()
extra_trees = ExtraTreesRegressor()
svr = SVR()
xgb = XGBRegressor()

# Create a list of tuples containing the model names and their instances
models = [
    ('linear_reg', linear_reg),
    ('Ridge', ridge),
    ('Lasso', lasso),
    ('KNeighborsRegressor', knn),
    ('DecisionTreeRegressor', decision_tree),
    ('RandomForestRegressor', random_forest),
    ('GradientBoostingRegressor', gradient_boosting),
    ('AdaBoostRegressor', adaboost),
    ('ExtraTreesRegressor', extra_trees),
    ('SVR', svr),
    ('XGBRegressor', xgb)
]
```

```python
# Initialize a list to store the results
results = []

# Train and evaluate each regression model individually
for name, model in models:
    pipeline = Pipeline([
        ('scaler', StandardScaler()),  # StandardScaler for feature scaling (optional)
        (name, RegressorTransformer(model)),
    ])

    # Train the pipeline on the dataset
    pipeline.fit(X_train, y_train)

    # Evaluate the performance of the regression model
    y_pred = pipeline.transform(X_test)
    mae = mean_absolute_error(y_test, y_pred)
    r_square = r2_score(y_test, y_pred)

     # Append the results to the list
    results.append({
        "Model": name,
        "mean_absolute_error (MAE)":mae,
        "R-squared (R2)": r_square*100
    })
    pass
```

```python
# Create a DataFrame from the results list
Algo_result = pd.DataFrame(results)
```

Why np.log?
> Nonlinear relationships: In some cases, the relationship between the predictor variable and the response variable may not be linear. Taking the logarithm of the predictor variable can help transform the relationship into a more linear form, making it easier to model using linear regression techniques.

> Skewed data: If the predictor variable is highly skewed or has a long tail, taking the logarithm can help normalize the distribution. This can be particularly useful when dealing with variables that have a wide range of values or exhibit exponential growth or decay

"Column transfer"?

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X1, y1, test_size = 0.2, random_state = 42)
```

```python
#using ct and one hot encoding
#decison tree
ct = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first',handle_unknown='ignore'),[0,1,5,6,11])
],remainder='passthrough')

mo = DecisionTreeRegressor(max_depth=8)

pipe = Pipeline([
    ('ct',ct),
    ('mo',mo)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

```python
#dRandomForestRegressor
ct = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first',handle_unknown='ignore'),[0,1,5,6,11])
],remainder='passthrough')

mo = RandomForestRegressor(n_estimators=100, random_state=3, max_samples=0.5,max_features=0.75,max_depth=15)

pipe = Pipeline([
    ('ct',ct),
    ('mo',mo)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```
